```python
#convert numeric columns to proper type first
data["ApplicantIncome"]=pd.to_numeric(data["ApplicantIncome"],errors="coerce")
data["Loan_Amount_Term"]=pd.to_numeric(data["Loan_Amount_Term"],errors="coerce")
data["Credit_History"]=pd.to_numeric(data["Credit_History"],errors="coerce")
data["Dependents"]=pd.to_numeric(data["Dependents"],errors="coerce")
data
```

```python
#fill missing numeric values with 0
data["Self_Employed"].fillna(0,inplace=True)
data["Credit_History"].fillna(data["Credit_History"].mean(),inplace=True)
data
```

```python
#1.bar graph
import matplotlib.pyplot as plt
#1.loan status count(approved vs not approved)("Loan Appilcation Status"):
plt.figure(figsize=(6,4))
data["Loan_Status"].value_counts().plot(kind="bar",color="pink")
plt.title("Loan Application Status")
plt.xlabel("status")
plt.ylabel("Number of Applicants")
plt.xticks(rotation=0)
plt.show()
```

```python
#2.loan amount distribution(histogram)
plt.figure(figsize=(6,4))
data["LoanAmount"].plot(kind="hist",bins=30,color="purple",edgecolor="black")
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#3.pie chart
import matplotlib.pyplot as plt
import seaborn as sns
#count loan status
Loan_Status_Counts=data["Loan_Status"].value_counts()
colors=sns.color_palette("pastel")[0:2]
#plot pie chart
plt.figure(figsize=(6,6,))
plt.pie(
    Loan_Status_Counts,
    labels=Loan_Status_Counts.index,
    autopct="%1.1f%%",
    colors=colors,
    startangle=90
    )
plt.title("Loan Application Status Distribution")
plt.show()
```

```python
#6.heatmap
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#correlation matrix
corr=data.corr(numeric_only=True)
#only numeric columns
#plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap Of Loan Data")
plt.show()
```

```python
#8.treemap
import plotly.express as px
import pandas as pd

# Sample loan dataset
data = {
    "LoanPurpose": ["Home Improvement", "Debt Consolidation", "Credit Card", "Business", "Other"],
    "LoanAmount": [150000, 400000, 250000, 100000, 50000]
}
df = pd.DataFrame(data)

# Create Treemap
fig = px.treemap(
    df,
    path=["LoanPurpose"],       # hierarchy
    values="LoanAmount",        # size of each block
    title="Loan Amount Distribution by Purpose"
)

fig.show()
```

```python
#9.waterfall chart
import pandas as pd
import plotly.graph_objects as go

# Example loan funnel data
stages = ["Total Applications", "Approved", "Funded", "Repaid", "Default"]
values = [1000000, -200000, -300000, -250000, -250000]

fig = go.Figure(go.Waterfall(
    name="Loan Flow",
    orientation="v",
    measure=["absolute", "relative", "relative", "relative", "relative"],
    x=stages,
    text=[f"${abs(v):,}" for v in values],
    y=values
))

fig.update_layout(
    title="Loan Waterfall Chart",
    waterfallgap=0.3
)
fig.show()
```

```python
#10.geographic map
import plotly.express as px
import pandas as pd

# Example loan data by US state
data = {
    "state": ["CA", "TX", "NY", "FL", "IL", "PA"],
    "loan_amount": [500000, 350000, 400000, 300000, 200000, 250000]
}
df = pd.DataFrame(data)

# Create Choropleth Map
fig = px.choropleth(
    df,
    locations="state",
    locationmode="USA-states",   # state-level mapping
    color="loan_amount",
    color_continuous_scale="Blues",
    scope="usa",
    title="Loan Amount Distribution by State"
)

fig.show()
```

```python
#11.network diagram
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add borrowers and loan purposes
borrowers = ["B1", "B2", "B3", "B4"]
purposes = ["Home Improvement", "Debt Consolidation", "Business"]

# Example relationships (borrower â†’ loan purpose)
edges = [
    ("B1", "Home Improvement"),
    ("B2", "Debt Consolidation"),
    ("B3", "Business"),
    ("B4", "Debt Consolidation"),
    ("B1", "Business")  # borrower may have multiple purposes
]

# Add edges
G.add_edges_from(edges)

# Draw network
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)  # layout
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=10, font_weight="bold")
plt.title("Loan Network Diagram (Borrowers â†’ Loan Purposes)")
plt.show()
```
